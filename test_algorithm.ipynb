```python
from algorithm_avl import generate_random_segments_avl, find_first_intersection_in_avl_tree
from algorithm_23 import generate_random_segments_23, find_first_intersection_in_23_tree
import matplotlib.pyplot as plt
import time
from datetime import datetime
```

```python
def plot_segments(segments, intersecting_pair=None):
    plt.figure(figsize=(8, 8))

    for segment in segments:
        x = [segment.left.x, segment.right.x]
        y = [segment.left.y, segment.right.y]
        plt.plot(x, y, marker='o', linestyle='-', color='blue')

    if intersecting_pair:
        x1 = [intersecting_pair[0].left.__dict__['x'],
             intersecting_pair[0].right.__dict__['x']]
        
        x2 = [intersecting_pair[1].left.__dict__['x'],
             intersecting_pair[1].right.__dict__['x']]
        
        y1 = [intersecting_pair[0].left.__dict__['y'],
             intersecting_pair[0].right.__dict__['y']]
        
        y2 = [intersecting_pair[1].left.__dict__['y'],
             intersecting_pair[1].right.__dict__['y']]

        plt.plot(x1, y1, marker='o', linestyle='-', color='red', label='Intersecting Pair 1')
        plt.plot(x2, y2, marker='o', linestyle='-', color='green', label='Intersecting Pair 2')

    plt.title('Random Segments')
    plt.xlabel('X-axis')
    plt.ylabel('Y-axis')
    plt.legend()
    plt.grid(True)
    plt.show()
```

## AVL-tree

```python
num_segments = 15  #     
x_range = (-20, 20)  #    x
y_range = (-20, 20)  #    y

segments = generate_random_segments_avl(num_segments, x_range, y_range)

print("Generated Segments:")
for idx, segment in enumerate(segments, start=1):
    print(f"Segment {idx}: Line - {segment.left.__dict__} to {segment.right.__dict__}")

start = time.perf_counter()
intersecting_pair = find_first_intersection_in_avl_tree(segments)
end = time.perf_counter()
print("Time: ", str(end - start))
plot_segments(segments, intersecting_pair)
```

```python
times1 = []
for i in range(1, 501):
    segments = generate_random_segments_avl(i*20, x_range, y_range)
    start = time.perf_counter()
    intersecting_pair = find_first_intersection_in_avl_tree(segments)
    end = time.perf_counter()
    res_time = end - start
    times1.append(res_time)
    print("Time: ", str(res_time))

num_segments1 = [i*20 for i in range(1, 501)]
```

```python
plt.plot(num_segments1, times1)
```

## Comparsion

```python
x_range = (-20, 20)  #    x
y_range = (-20, 20)  #    y

times_avl, times_23 = [], []
for i in range(1, 501):
    segments = generate_random_segments_23(i*20, x_range, y_range)

    start1 = time.perf_counter()
    intersecting_pair1 = find_first_intersection_in_avl_tree(segments)
    end1 = time.perf_counter()
    res_time1 = end1 - start1

    start2 = time.perf_counter()
    intersecting_pair2 = find_first_intersection_in_23_tree(segments)
    end2 = time.perf_counter()
    res_time2 = end2 - start2
    
    times_avl.append(res_time1)
    times_23.append(res_time2)
    #print("Time: ", str(res_time))

num_segments = [i*20 for i in range(1, 501)]
```

```python
plt.figure(figsize=(15,10))
plt.plot(num_segments, times_avl, label="AVL tree")
plt.plot(num_segments, times_23, label="2-3 tree")
plt.legend()
```
